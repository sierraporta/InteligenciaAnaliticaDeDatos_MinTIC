```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv("FAOSTAT_data_en_7-6-2023 (1).csv")
```

```python
countries=data["Area"].unique()
item=data["Item"].unique()
```

```python
data2=data.pivot(index='Area', columns='Item', values='Value')
```

```python
data2.columns
```

```python
data2.info()
```

```python
data20=data2[['Animal fats', 'Cereals - Excluding Beer', 'Eggs', 'Fish, Seafood',
       'Fruits - Excluding Wine', 'Meat', 'Milk - Excluding Butter',
       'Offals', 'Oilcrops', 'Starchy Roots', 'Vegetable Oils', 'Vegetables']]
```

```python
data20
```

```python
data20.info()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline 
import os
import warnings
warnings.filterwarnings('ignore')
```

```python
X=data20.dropna()
from sklearn.preprocessing import MinMaxScaler, StandardScaler
scaler = StandardScaler()
X=pd.DataFrame(scaler.fit_transform(X),columns=X.columns)

X=X[(X["Meat"]<2)&(X["Vegetables"]<1)]

X0=X.copy()
```

```python
from sklearn.cluster import KMeans

clusters = []
for i in range(1, 11):
    km = KMeans(n_clusters=i).fit(X)
    clusters.append(km.inertia_)
    
fig, ax = plt.subplots(figsize=(8, 6))
sns.lineplot(x=list(range(1, 11)), y=clusters, ax=ax)
ax.set_title('Searching for Elbow')
ax.set_xlabel('Clusters')
ax.set_ylabel('Inertia')
plt.show()
```

```python
# 2 cluster
km2 = KMeans(n_clusters=2).fit(X0)
X['Labels_k2'] = km2.labels_
# 3 cluster
km3 = KMeans(n_clusters=3).fit(X0)
X['Labels_k3'] = km3.labels_
# 4 cluster
km4 = KMeans(n_clusters=4).fit(X0)
X['Labels_k4'] = km4.labels_
# 5 cluster
km5 = KMeans(n_clusters=5).fit(X0)
X['Labels_k5'] = km5.labels_
# 6 cluster
km6 = KMeans(n_clusters=6).fit(X0)
X['Labels_k6'] = km6.labels_
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=X, x="Meat", y="Vegetables", hue='Labels_k6', palette=sns.color_palette('tab10'), s=20)
plt.title('KMeans with 3 Clusters')
plt.show()
```
