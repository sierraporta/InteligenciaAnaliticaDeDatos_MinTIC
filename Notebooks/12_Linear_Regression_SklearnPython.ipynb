```python
# 1. Lo primero es tener los datos
datos=pd.read_csv("Life Expectancy Data.csv")
```

```python
# Saber que tipo de variables tengo
datos.info()
```

```python
datos.describe().T
```

```python
## Eliminando los datos missing...
datos=datos.dropna()
```

```python
datos
```

```python
# Eliminando las columnas no num√©ricas
datos=datos.drop(columns=["Country","Status"])
datos
```

```python
# 2. Crear un modelo, machine learning, Nombre: Regresion Lineal M√∫ltiple
# 2.1. Separar datos en variables que predicen (predictandos) y variable a predecir (predicci√≥n)
X=datos.drop(columns=["Life expectancy "]) # Variables predictoras
y=datos["Life expectancy "] # Variable a predecir
```

```python
y
```

```python
# 2.2. Segmentar el conjunto de datos en Entrenamiento y Prueba (Dividir)
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=9860967)
```

```python
y_train
```

```python
X_train
```

```python
# 2.3. Defininiendo el modelo, es "regresi√≥n lineal m√∫ltiple"
# Cargamos el modelo
### Llamamos la librer√≠a de machine learning de regresiones lineales
from sklearn import linear_model 
david = linear_model.LinearRegression() #Este es el modelo y se llama "david"
```

```python
david
```

```python
# 2.4. El modelo se entrena con los datos de entrada, en este caso los datos de entrenamiento...
david.fit(X_train,y_train)
```

```python
y_train
```

```python
# 2.5. Predicciones
david.predict(X_train)
```

```python
## Predicci√≥n sobre el conjunto de prueba
david.predict(X_test)
```

```python
y_test
```

```python
### Evaluar el modelo, evaluar la predicci√≥n....
#### M√©tricas: Error cuadr√°tico medio, desviaci√≥n est√°ndar
Error=np.sqrt(np.mean((y_test-david.predict(X_test))**2))
```
