```python
#2. Imputación
data2=data.drop(columns=["Country","Year","Status"]).copy()
```

```python
data2["Life expectancy "]=data2["Life expectancy "].fillna(data2["Life expectancy "].median())
```

```python
for i in data2.columns:
    data2[i]=data2[i].fillna(data2[i].mean())
```

In reality, a relationship may exist between the dependent variable and multiple independent variables. For these types of models (assuming linearity), we can use Multiple Linear Regression with the following structure:

Y = C + M1*X1 + M2*X2 + …

Vamos a construir el modelo lineal...

```python
import statsmodels.api as smod
```

Haremos dos modelos, uno para data1, y uno para data2

```python
#Modelo1 para data1
```

```python
X1=data1.drop(columns=["Life expectancy "])
Y1=data1["Life expectancy "]
```

```python
X1=smod.add_constant(X1)
```

```python
modelo1=smod.OLS(Y1,X1).fit()
```

```python
modelo1.summary()
```

**Variables que no explican (modelo1):**
- Alcohol

- Polio
- Total expenditure
- GDP
- Population
- thinness 1-19 years
- thinness 5-9 years

```python
X2=data2.drop(columns=["Life expectancy "])
Y2=data2["Life expectancy "]
```

```python
X2=smod.add_constant(X2)
```

```python
modelo2=smod.OLS(Y2,X2).fit()
```

```python
modelo2.summary()
```

```python
predicciones1=modelo1.predict(X1)
predicciones2=modelo2.predict(X2)
```
