# ¿Qué es la regresión Gradient Boosting?
El algoritmo Gradient Boosting se utiliza para generar un modelo ensemble combinando los aprendices débiles o los modelos predictivos débiles. El algoritmo Gradient Boosting puede utilizarse para entrenar modelos tanto para problemas de regresión como de clasificación. El algoritmo Gradient Boosting Regression se utiliza para ajustar el modelo que predice el valor continuo.

Gradient boosting construye un modo aditivo utilizando múltiples árboles de decisión de tamaño fijo como aprendices débiles o modelos predictivos débiles. El parámetro n_estimators decide el número de árboles de decisión que se utilizarán en las etapas de boosting. Gradient Boost se diferencia de AdaBoost en que en AdaBoost se utilizan tocones de decisión (un nodo y dos hojas), mientras que en Gradient Boosting se utilizan árboles de decisión de tamaño fijo.

El proceso de ajuste del modelo comienza con una constante, como el valor medio de los valores objetivo. En las etapas siguientes, los árboles de decisión o los estimadores se ajustan para predecir los gradientes negativos de las muestras. Los gradientes se actualizan en cada iterador (para cada estimador posterior). Se utiliza una tasa de aprendizaje para reducir el resultado o la contribución de cada uno de los árboles o estimadores posteriores.

El modelo de regresión Gradient Boosting crea un bosque de 1000 árboles con una profundidad máxima de 3 y una pérdida mínima cuadrática.
Los hiperparámetros utilizados para el entrenamiento de los modelos son los siguientes
- n_estimadores: Número de árboles utilizados para el boosting
- profundidad_máx: Profundidad máxima del árbol
- learning_rate: Tasa a la que se escalarán o reducirán los resultados de cada árbol
- pérdida: Función de pérdida a optimizar. Las opciones para las funciones de pérdida son: ls, lad, huber, quantile. ls representa la pérdida mínima cuadrática. lad representa la desviación mínima absoluta. huber representa la combinación de ambas, ls y lad.
- El coeficiente de determinación, R-cuadrado, se utiliza para medir la precisión del modelo. La puntuación del método invocada en la instancia de regresión de aumento de gradiente daría como resultado la impresión del R-cuadrado

# Manos a la obra

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn import datasets
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.pipeline import make_pipeline
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.decomposition import PCA
from sklearn.metrics import mean_squared_error
#
# Load the Boston Dataset
#
bhp = pd.read_csv("FoodAppBusiness.csv")
```

```python
bhp
```

```python
X=bhp.drop(columns=["TotalNoOfCampaignAccepted"])
y=bhp["TotalNoOfCampaignAccepted"]
#
# Create Training and Test Split
#
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, test_size=0.1)
#
# Standardize the dataset
#
sc = StandardScaler()
X_train_std = sc.fit_transform(X_train)
X_test_std = sc.transform(X_test)
#
# Hyperparameters for GradientBoostingRegressor
#
gbr_params = {'n_estimators': 1000,
          'max_depth': 3,
          'min_samples_split': 5,
          'learning_rate': 0.01,
          'loss': 'quantile'}
#
# Create an instance of gradient boosting regressor
#
gbr = GradientBoostingRegressor(**gbr_params)
#
# Fit the model
#
gbr.fit(X_train_std, y_train)
#
# Print Coefficient of determination R^2
#
print("Model Accuracy: %.3f" % gbr.score(X_test_std, y_test))
#
# Create the mean squared error
#
mse = mean_squared_error(y_test, gbr.predict(X_test_std))
print("The mean squared error (MSE) on test set: {:.4f}".format(mse))
```

## Determinar la importancia de las características
Este es el código para determinar la importancia de una característica. Observe el uso del atributo feature_importances_ para calcular los valores de importancia de las características.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.inspection import permutation_importance
#
# Get Feature importance data using feature_importances_ attribute
#
feature_importance = gbr.feature_importances_
sorted_idx = np.argsort(feature_importance)
pos = np.arange(sorted_idx.shape[0]) + .5
fig = plt.figure(figsize=(8, 8))
plt.barh(pos, feature_importance[sorted_idx], align='center')
plt.yticks(pos, np.array(bhp.columns)[sorted_idx])
plt.title('Feature Importance (MDI)')
result = permutation_importance(gbr, X_test_std, y_test, n_repeats=10,
                                random_state=42, n_jobs=2)
sorted_idx = result.importances_mean.argsort()
fig.tight_layout()
plt.show()
```
