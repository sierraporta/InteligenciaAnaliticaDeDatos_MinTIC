### Visualizando datos

```python
import matplotlib.pyplot as mpl
```

```python
mpl.figure(figsize=(20,5))
mpl.plot(df_ex["Date"],df_ex["National-US"],color="black",linewidth=3,marker="o",markersize=0, label="US house price Index (Case-shiller)")
mpl.plot(df_ex["Date"],df_ex["FL-Miami"],color="blue",linewidth=3,marker="o",markersize=0, label="Florida-Miami house price Index (Case-shiller)")
mpl.plot(df_ex["Date"],df_ex['DC-Washington'],color="red",linewidth=3,marker="o",markersize=0, label='DC-Washington house price Index (Case-shiller)')
mpl.plot(df_ex["Date"],df_ex['MI-Detroit'],color="green",linewidth=3,marker="o",markersize=0, label='MI-Detroit house price Index (Case-shiller)')

mpl.xlabel("Date", fontsize=16)
mpl.ylabel("National-US", fontsize=16)
mpl.xticks(fontsize=14)
mpl.yticks(fontsize=14)
mpl.grid(alpha=0.3)
mpl.legend(loc=2, fontsize=16)
#mpl.title("US house price Index (Case-shiller)", fontsize=16, fontweight="bold", color="b")
mpl.text(df_ex["Date"][250],30,"Fuente de los datos: https://datahub.io/core/house-prices-us")
mpl.show()
```
