```python
#¿Cuántos países están en el rango de felicidad entre 4.28 y 6.5?
datos2018=datos[datos["Year"]==2018]
datos2019=datos[datos["Year"]==2019]
N2018=datos[(datos["Score"]>=4.28) & (datos["Score"]<=6.5) & (datos["Year"]==2018)]
N2019=datos[(datos["Score"]>=4.28) & (datos["Score"]<=6.5) & (datos["Year"]==2019)]
```

```python
datos2018.describe().T
```

```python
datos2019.describe().T
```

```python
print("En el año 2018 hay ",len(N2018), "países, mientras que en el año 2019 hay", len(N2019), "países")
```

```python
#Porcentajes del total:
print("En el año 2018 hay ",np.round(len(N2018)/len(datos2018)*100,1),"%", "países, mientras que en el año 2019 hay", 
      np.round(len(N2019)/len(datos2019)*100,1),"%", "países")
```

# Estadísticos descriptivos:
- media
- mediana
- moda
- varianza
- desviación estándar
- coeficiente de fisher (asimetría)
- curtosis

```python
#Media
datos2018["Score"].mean(), np.mean(datos2018["Score"]), datos2018["Score"].mean(), np.mean(datos2019["Score"])
```

```python
datos2018.mean()
```

```python
datos.columns
```

```python
datos2018_r=datos2018.drop(columns=['Country or region','Overall rank','Year'])
datos2019_r=datos2019.drop(columns=['Country or region','Overall rank','Year'])
```

```python
datos2018_r
```

```python
# forma de seleccionar solo columnas numéricas...
datos2018.select_dtypes(include=np.number)
```

```python
datos2018_r.mean()
```

```python
datos2019_r.mean()
```

```python
datos2018_r.median()
```

```python
datos2018_r.mode()
```

```python
#Calculando la varianza y la desviación estándar
datos2018_r.var(), datos2018_r.std(), np.sqrt(datos2018_r.var())
```

```python
# Calculando la asimetría o coeficiente de Fisher..
datos2018_r.skew()
```

```python
#Calculando la curtosis...
datos2018_r.kurtosis()
```

```python
datos2019_r.kurtosis()
```

```python
import matplotlib.pyplot as mpl
```

```python
mpl.figure(figsize=(20,8))
datos2018_r.hist(figsize=(20,10),bins=20,color="red",histtype="step",linewidth=3)
datos2019_r.hist(figsize=(20,10),bins=20,color="blue",histtype="step",linewidth=3)
mpl.legend()
mpl.show()
```

```python
mpl.figure(figsize=(8,6))
datos2018_r["Perceptions of corruption"].hist(figsize=(6,4),bins=20,color="red",histtype="stepfilled",linewidth=3,label="2018",alpha=0.5)
datos2019_r["Perceptions of corruption"].hist(figsize=(6,4),bins=20,color="blue",histtype="stepfilled",linewidth=3,label="2019",alpha=0.3)
mpl.legend()
mpl.xlabel("Perceptions of corruption"); mpl.ylabel("Número de países")
mpl.show()
```

```python
mpl.figure(figsize=(8,6))
datos2018_r["Perceptions of corruption"].hist(figsize=(6,4),bins=20,color="red",histtype="stepfilled",linewidth=3,label="2018",alpha=0.5,density=True)
datos2019_r["Perceptions of corruption"].hist(figsize=(6,4),bins=20,color="blue",histtype="stepfilled",linewidth=3,label="2019",alpha=0.3,density=True)
mpl.legend()
mpl.xlabel("Perceptions of corruption"); mpl.ylabel("Número de países")
mpl.show()
```

```python
mpl.figure(figsize=(8,6))
datos2018_r["Perceptions of corruption"].hist(figsize=(6,4),bins=20,color="red",histtype="stepfilled",linewidth=3,label="2018",alpha=0.5,density=True,cumulative=True)
datos2019_r["Perceptions of corruption"].hist(figsize=(6,4),bins=20,color="blue",histtype="stepfilled",linewidth=3,label="2019",alpha=0.3,density=True,cumulative=True)
mpl.legend(loc=2)
mpl.xlabel("Perceptions of corruption"); mpl.ylabel("Número de países")
mpl.show()
```
