# Análisis de series temporales en Python

## ¿Qué es el análisis de series temporales?
A veces los datos cambian con el tiempo. Estos datos se denominan datos dependientes del tiempo. Con datos dependientes del tiempo, se puede analizar el pasado para predecir el futuro. La predicción futura también incluirá el tiempo como variable, y el resultado variará con el tiempo. Utilizando datos dependientes del tiempo, puede encontrar patrones que se repiten a lo largo del tiempo.

Una serie temporal es un conjunto de observaciones que se recogen tras intervalos regulares de tiempo. Si se grafican, las Series de Tiempo siempre tendrían uno de sus ejes como tiempo.
![](https://www.dataquest.io/wp-content/uploads/2018/11/time-series-pandas_66_0.png)

Consideremos el funcionamiento de una panadería. Dados los datos de los últimos meses, puede predecir qué artículos necesita hornear y a qué hora. La gente de la mañana necesitaría más artículos de pan, como panecillos, cruasanes, magdalenas para el desayuno, etc. Por la noche, la gente puede venir a comprar pasteles y otros postres. El análisis de series temporales permite predecir la demanda de productos en diferentes momentos e incluso en diferentes estaciones.

## ¿Cuáles son los distintos componentes del análisis de series temporales?
El siguiente diagrama muestra los diferentes componentes del Análisis de Series Temporales:

### Tendencia: 
La Tendencia muestra la variación de los datos con el tiempo o la frecuencia de los datos. Utilizando una Tendencia, puede ver cómo sus datos aumentan o disminuyen con el tiempo. Los datos pueden aumentar, disminuir o permanecer estables. A lo largo del tiempo, la población, las fluctuaciones bursátiles y la producción de una empresa son ejemplos de tendencias.
### Estacionalidad: 
La estacionalidad se utiliza para encontrar las variaciones que se producen a intervalos regulares de tiempo. Algunos ejemplos son los festivales, las convenciones, las estaciones, etc. Estas variaciones suelen producirse en torno al mismo periodo de tiempo y afectan a los datos de formas específicas que se pueden predecir. 
### Irregularidad: 
Las fluctuaciones en los datos de las series temporales no se corresponden con la tendencia o la estacionalidad. Estas variaciones en las series temporales son puramente aleatorias y suelen estar causadas por circunstancias imprevisibles, como un descenso repentino de la población debido a una calamidad natural.
### Cíclicas: 
Las oscilaciones en las series temporales que duran más de un año se denominan cíclicas. Pueden ser periódicas o no. 
### Estacionaria: 
Una serie temporal que tiene las mismas propiedades estadísticas a lo largo del tiempo es estacionaria. Las propiedades siguen siendo las mismas en cualquier punto de la serie. Los datos deben ser estacionarios para poder realizar análisis de series temporales con ellos. Una serie estacionaria tiene una media, varianza y covarianza constantes.

## Modelo ARIMA
El modelo ARIMA significa media móvil integrada autorregresiva. Se utiliza para predecir los valores futuros de una serie temporal utilizando sus valores pasados y los errores de previsión. El siguiente diagrama muestra los componentes de un modelo ARIMA: 

![](https://miro.medium.com/v2/resize:fit:640/format:webp/1*JrxZNAcWhDUHud0QPvYP8A.png)

### Modelo autorregresivo
Los modelos autorregresivos predicen el comportamiento futuro utilizando el comportamiento pasado cuando existe cierta correlación entre los datos pasados y futuros. La fórmula siguiente representa el modelo autorregresivo. Se trata de una versión modificada de la fórmula de la pendiente en la que el valor objetivo se expresa como la suma del intercepto, el producto de un coeficiente y la salida anterior, y un término de corrección de errores.

![](https://miro.medium.com/v2/resize:fit:720/format:webp/1*uaWPoUH1V9z_N4EUWK7edQ.jpeg)

### Media móvil
La media móvil es un método estadístico que toma la media actualizada de los valores para ayudar a reducir el ruido. Toma el promedio sobre un intervalo de tiempo específico. Se puede obtener tomando diferentes subconjuntos de datos y hallando sus respectivas medias.

Primero se considera un grupo de puntos de datos y se calcula su media. Luego encuentras el siguiente promedio eliminando el primer valor de los datos e incluyendo el siguiente valor de la serie.

### Integración
La integración es la diferencia entre las observaciones actuales y las anteriores. Se utiliza para hacer estacionaria la serie temporal. 

Cada uno de estos valores actúa como un parámetro para nuestro modelo ARIMA. En lugar de representar el modelo ARIMA mediante estos distintos operadores y modelos, se utilizan parámetros para representarlos. Estos parámetros son: 

p: Valores retardados anteriores para cada punto temporal. Derivado del modelo autorregresivo.
q: Valores retardados anteriores del término de error. Derivado de la media móvil.
d: Número de veces que se diferencian los datos para hacerlos estacionarios. Es el número de veces que realiza la integración.

## Estructuras de datos de series temporales
Antes de sumergirnos en los datos OPSD, vamos a introducir brevemente las principales estructuras de datos de pandas para trabajar con fechas y horas. En pandas, un único punto en el tiempo se representa como un Timestamp. Podemos utilizar la función to_datetime() para crear Timestamps a partir de cadenas en una amplia variedad de formatos de fecha/hora. Vamos a importar pandas y convertir algunas fechas y horas en Timestamps.

```python
import pandas as pd
pd.to_datetime('2018-01-15 3:45pm')
```

```python
pd.to_datetime('7/8/1952')
```

Como podemos ver, to_datetime() infiere automáticamente un formato de fecha/hora basándose en la entrada. En el ejemplo anterior, la fecha ambigua '7/8/1952' se asume como mes/día/año y se interpreta como 8 de julio de 1952. Alternativamente, podemos usar el parámetro dayfirst para decirle a pandas que interprete la fecha como 7 de agosto de 1952.

```python
pd.to_datetime('7/8/1952', dayfirst=True)
```

Si proporcionamos una lista o array de cadenas como entrada a to_datetime(), devuelve una secuencia de valores fecha/hora en un objeto DatetimeIndex, que es la estructura de datos central que potencia gran parte de la funcionalidad de series temporales de pandas.

```python
pd.to_datetime(['2018-01-05', '7/8/1952', 'Oct 10, 1995'])
```

En el DatetimeIndex anterior, el tipo de datos datetime64[ns] indica que los datos subyacentes se almacenan como enteros de 64 bits, en unidades de nanosegundos (ns). Esta estructura de datos permite a pandas almacenar de forma compacta grandes secuencias de valores fecha/hora y realizar eficientemente operaciones vectorizadas usando arrays NumPy datetime64.

Si estamos tratando con una secuencia de cadenas todas en el mismo formato de fecha/hora, podemos especificarlo explícitamente con el parámetro format. Para conjuntos de datos muy grandes, esto puede acelerar enormemente el rendimiento de to_datetime() en comparación con el comportamiento por defecto, donde el formato se infiere por separado para cada cadena individual. Se puede utilizar cualquiera de los códigos de formato de las funciones strftime() y strptime() del módulo datetime incorporado en Python. El ejemplo siguiente utiliza los códigos de formato %m (mes numérico), %d (día del mes) y %y (año de 2 dígitos) para especificar el formato.

Además de los objetos Timestamp y DatetimeIndex que representan puntos individuales en el tiempo, pandas también incluye estructuras de datos que representan duraciones (por ejemplo, 125 segundos) y periodos (por ejemplo, el mes de noviembre de 2018). Para más información sobre estas estructuras de datos, hay un buen resumen aquí. En este tutorial utilizaremos DatetimeIndexes, la estructura de datos más común para series temporales de pandas.

```python
pd.to_datetime(['2018-01-05', '7/8/1952', 'Oct 10, 1995'],format='mixed')
```

## Indexación temporal
Una de las características más potentes y convenientes de las series temporales de pandas es la indexación basada en el tiempo - utilizando fechas y horas para organizar y acceder intuitivamente a nuestros datos. Con la indexación basada en el tiempo, podemos utilizar cadenas formateadas en fecha/hora para seleccionar datos en nuestro DataFrame con el accesor loc. La indexación funciona de forma similar a la indexación estándar basada en etiquetas con loc, pero con algunas características adicionales.

Por ejemplo, podemos seleccionar los datos de un solo día utilizando una cadena como '2017-08-10'.

```python
opsd_daily.loc['2017-08-10']
```

También podemos seleccionar un tramo de días, como '2014-01-20':'2014-01-22'. Al igual que con la indexación normal basada en etiquetas con loc, la porción incluye ambos extremos.

```python
opsd_daily.loc['2014-01-20':'2014-01-22']
```

Otra característica muy útil de las series temporales de pandas es la indexación de cadenas parciales, donde podemos seleccionar todas las fechas/horas que coincidan parcialmente con una cadena dada. Por ejemplo, podemos seleccionar todo el año 2006 con opsd_daily.loc['2006'], o todo el mes de febrero de 2012 con opsd_daily.loc['2012-02'].

```python
opsd_daily.loc['2012-02']
```

## Personalización de los gráficos de series temporales
Para visualizar mejor la estacionalidad semanal del consumo eléctrico en el gráfico anterior, estaría bien disponer de líneas de cuadrícula verticales en una escala temporal semanal (en lugar de en el primer día de cada mes). Podemos personalizar nuestro gráfico con matplotlib.dates, así que vamos a importar ese módulo.

```python
import matplotlib.dates as mdates
```

Como las marcas de fecha/hora se manejan de forma un poco diferente en matplotlib.dates en comparación con el método plot() de DataFrame, vamos a crear el gráfico directamente en matplotlib. A continuación, utilizamos mdates.WeekdayLocator() y mdates.MONDAY para establecer las marcas del eje x en el primer lunes de cada semana. También usamos mdates.DateFormatter() para mejorar el formato de las etiquetas de las marcas, usando los códigos de formato que vimos antes.

```python
fig, ax = plt.subplots()
ax.plot(opsd_daily.loc['2017-01':'2017-02', 'Consumption'], marker='o', linestyle='-')
ax.set_ylabel('Daily Consumption (GWh)')
ax.set_title('Jan-Feb 2017 Electricity Consumption')
# Set x-axis major ticks to weekly interval, on Mondays
ax.xaxis.set_major_locator(mdates.WeekdayLocator(byweekday=mdates.MONDAY))
# Format x-tick labels as 3-letter month name and day number
ax.xaxis.set_major_formatter(mdates.DateFormatter('%b %d'));
```

Ahora tenemos líneas de cuadrícula verticales y etiquetas con un bonito formato en cada lunes, de forma que podemos saber fácilmente qué días son laborables y qué días son de fin de semana.

## Estacionalidad
A continuación, vamos a explorar más a fondo la estacionalidad de nuestros datos con gráficos de caja, utilizando la función boxplot() de seaborn para agrupar los datos por diferentes períodos de tiempo y mostrar las distribuciones de cada grupo. Primero agruparemos los datos por meses, para visualizar la estacionalidad anual.

```python
fig, axes = plt.subplots(3, 1, figsize=(11, 10), sharex=True)
for name, ax in zip(['Consumption', 'Solar', 'Wind'], axes):
    sns.boxplot(data=opsd_daily, x='Month', y=name, ax=ax)
    ax.set_ylabel('GWh')
    ax.set_title(name)
    # Remove the automatic x-axis label from all but the bottom subplot
    if ax != axes[-1]:
        ax.set_xlabel('')
```

Estos gráficos de caja confirman la estacionalidad anual que vimos en gráficos anteriores y aportan algunos datos adicionales:

* Aunque el consumo de electricidad suele ser mayor en invierno y menor en verano, la mediana y los dos cuartiles inferiores son más bajos en diciembre y enero en comparación con noviembre y febrero, probablemente debido a que las empresas cierran durante las vacaciones. Lo vimos en la serie temporal del año 2017, y el gráfico de cajas confirma que se trata de un patrón constante a lo largo de los años.
* Mientras que la producción de energía solar y eólica muestran una estacionalidad anual, las distribuciones de energía eólica tienen muchos más valores atípicos, lo que refleja los efectos de las velocidades de viento extremas ocasionales asociadas con tormentas y otras condiciones meteorológicas transitorias.

## Frecuencias
Cuando los puntos de datos de una serie temporal están espaciados uniformemente en el tiempo (por ejemplo, cada hora, cada día, cada mes, etc.), la serie temporal se puede asociar con una frecuencia en pandas. Por ejemplo, utilicemos la función date_range() para crear una secuencia de fechas espaciadas uniformemente desde 1998-03-10 hasta 1998-03-15 con frecuencia diaria.

```python
pd.date_range('1998-03-10', '1998-03-15', freq='D')
```

El DatetimeIndex resultante tiene un atributo freq con valor 'D', que indica frecuencia diaria. Las frecuencias disponibles en pandas incluyen horaria ('H'), diaria de calendario ('D'), diaria de negocio ('B'), semanal ('W'), mensual ('M'), trimestral ('Q'), anual ('A'), y muchas otras. Las frecuencias también pueden especificarse como múltiplos de cualquiera de las frecuencias base, por ejemplo "5D" para cada cinco días.

Como otro ejemplo, creemos un intervalo de fechas con frecuencia horaria, especificando la fecha de inicio y el número de períodos, en lugar de la fecha de inicio y la fecha final.

```python
pd.date_range('2004-09-20', periods=8, freq='H')
```

Ahora echemos otro vistazo al DatetimeIndex de nuestra serie temporal opsd_daily.

```python
opsd_daily.index
```

Podemos ver que no tiene frecuencia (freq=None). Esto tiene sentido, ya que el índice se creó a partir de una secuencia de fechas en nuestro archivo CSV, sin especificar explícitamente ninguna frecuencia para la serie temporal.

Si sabemos que nuestros datos deben tener una frecuencia específica, podemos utilizar el método asfreq() del DataFrame para asignar una frecuencia. Si falta alguna fecha/hora en los datos, se añadirán nuevas filas para esas fechas/horas, que estarán vacías (NaN), o rellenadas según un método de relleno de datos especificado, como el relleno hacia delante o la interpolación.

Para ver cómo funciona esto, creemos un nuevo DataFrame que contenga sólo los datos de Consumo para el 3, 6 y 8 de febrero de 2013.

```python
# To select an arbitrary sequence of date/time values from a pandas time series,
# we need to use a DatetimeIndex, rather than simply a list of date/time strings
times_sample = pd.to_datetime(['2013-02-03', '2013-02-06', '2013-02-08'])
# Select the specified dates and just the Consumption column
consum_sample = opsd_daily.loc[times_sample, ['Consumption']].copy()
consum_sample
```

Ahora utilizamos el método asfreq() para convertir el DataFrame a frecuencia diaria, con una columna para los datos no rellenados, y una columna para los datos rellenados hacia adelante.

```python
# Convert the data to daily frequency, without filling any missings
consum_freq = consum_sample.asfreq('D')
# Create a column with missings forward filled
consum_freq['Consumption - Forward Fill'] = consum_sample.asfreq('D', method='ffill')
consum_freq
```

En la columna Consumo, tenemos los datos originales, con un valor de NaN para cualquier fecha que falte en nuestro DataFrame consum_sample. En la columna Consumo - Relleno hacia delante, los datos que faltan se han rellenado hacia delante, lo que significa que el último valor se repite a través de las filas que faltan hasta que aparece el siguiente valor que no falta.

Si está realizando cualquier análisis de series temporales que requiera datos espaciados uniformemente sin ninguna falta, querrá utilizar asfreq() para convertir su serie temporal a la frecuencia especificada y rellenar cualquier falta con un método apropiado.

## Remuestreo
A menudo es útil remuestrear nuestros datos de series temporales a una frecuencia más baja o más alta. El remuestreo a una frecuencia inferior (downsampling) suele implicar una operación de agregación, por ejemplo, calcular totales de ventas mensuales a partir de datos diarios. Los datos diarios de OPSD con los que estamos trabajando en este tutorial se redujeron a partir de las series temporales horarias originales. El remuestreo a una frecuencia más alta (remuestreo ascendente) es menos común y a menudo implica interpolación u otro método de relleno de datos - por ejemplo, interpolar datos meteorológicos horarios a intervalos de 10 minutos para introducirlos en un modelo científico.

Nos centraremos aquí en el muestreo descendente, explorando cómo puede ayudarnos a analizar nuestros datos OPSD en varias escalas de tiempo. Usamos el método resample() de DataFrame, que divide el DatetimeIndex en intervalos de tiempo y agrupa los datos por intervalos de tiempo. El método resample() devuelve un objeto Resampler, similar a un objeto pandas GroupBy. A continuación, podemos aplicar un método de agregación como mean(), median(), sum(), etc., al grupo de datos para cada intervalo de tiempo.

Por ejemplo, vamos a remuestrear los datos a una media semanal de series temporales.

```python
# Specify the data columns we want to include (i.e. exclude Year, Month, Weekday Name)
data_columns = ['Consumption', 'Wind', 'Solar', 'Wind+Solar']
# Resample to weekly frequency, aggregating with mean
opsd_weekly_mean = opsd_daily[data_columns].resample('W').mean()
opsd_weekly_mean.head(3)
```

La primera fila, denominada 2006-01-01, contiene la media de todos los datos contenidos en el intervalo 2006-01-01 a 2006-01-07. La segunda fila, denominada 2006-01-08, contiene la media de todos los datos contenidos en el intervalo 2006-01-08 a 2006-01-14, y así sucesivamente. La segunda fila, denominada 2006-01-08, contiene la media de los datos del intervalo 2006-01-08 a 2006-01-14, y así sucesivamente. Por defecto, cada fila de la serie temporal de muestreo descendente está etiquetada con el borde derecho del intervalo temporal.

Por construcción, nuestra serie temporal semanal tiene 1/7 de puntos de datos que la serie temporal diaria. Podemos confirmarlo comparando el número de filas de los dos DataFrames.

```python
print(opsd_daily.shape[0])
print(opsd_weekly_mean.shape[0])
```

Comparemos las series temporales diarias y semanales de Solar a lo largo de un semestre.

```python
# Start and end of the date range to extract
start, end = '2017-01', '2017-06'
# Plot daily and weekly resampled time series together
fig, ax = plt.subplots()
ax.plot(opsd_daily.loc[start:end, 'Solar'],
marker='.', linestyle='-', linewidth=0.5, label='Daily')
ax.plot(opsd_weekly_mean.loc[start:end, 'Solar'],
marker='o', markersize=8, linestyle='-', label='Weekly Mean Resample')
ax.set_ylabel('Solar Production (GWh)')
ax.legend();
```

Podemos ver que la serie temporal media semanal es más suave que la serie temporal diaria porque la variabilidad de mayor frecuencia se ha promediado en el remuestreo.

Ahora vamos a remuestrear los datos a la frecuencia mensual, agregando con sumas totales en lugar de la media. A diferencia de la agregación con mean(), que establece la salida a NaN para cualquier período con todos los datos que faltan, el comportamiento predeterminado de sum() devolverá la salida de 0 como la suma de los datos que faltan. Utilizamos el parámetro min_count para cambiar este comportamiento.

```python
# Compute the monthly sums, setting the value to NaN for any month which has
# fewer than 28 days of data
opsd_monthly = opsd_daily[data_columns].resample('M').sum(min_count=28)
opsd_monthly.head(3)
```

Puede observar que los datos remuestreados mensuales se etiquetan con el final de cada mes (el borde derecho de la bandeja), mientras que los datos remuestreados semanales se etiquetan con el borde izquierdo de la bandeja. Por defecto, los datos remuestreados se etiquetan con el borde derecho para las frecuencias mensuales, trimestrales y anuales, y con el borde izquierdo para el resto de frecuencias. Este comportamiento y otras opciones pueden ajustarse utilizando los parámetros enumerados en la documentación de resample().

Ahora vamos a explorar las series temporales mensuales trazando el consumo de electricidad como un gráfico de líneas, y la producción de energía eólica y solar juntas como un gráfico de áreas apiladas.

```python
fig, ax = plt.subplots()
ax.plot(opsd_monthly['Consumption'], color='black', label='Consumption')
opsd_monthly[['Wind', 'Solar']].plot.area(ax=ax, linewidth=0)
ax.xaxis.set_major_locator(mdates.YearLocator())
ax.legend()
ax.set_ylabel('Monthly Total (GWh)');
```

A esta escala temporal mensual, podemos ver claramente la estacionalidad anual en cada serie temporal, y también es evidente que el consumo de electricidad se ha mantenido bastante estable a lo largo del tiempo, mientras que la producción de energía eólica ha crecido constantemente, con la energía eólica + solar comprendiendo una parte cada vez mayor de la electricidad consumida.

Exploremos esto más a fondo volviendo a muestrear a frecuencia anual y calculando la relación entre eólica+solar y consumo para cada año.

```python
# Compute the annual sums, setting the value to NaN for any year which has
# fewer than 360 days of data
opsd_annual = opsd_daily[data_columns].resample('A').sum(min_count=360)
# The default index of the resampled DataFrame is the last day of each year,
# ('2006-12-31', '2007-12-31', etc.) so to make life easier, set the index
# to the year component
opsd_annual = opsd_annual.set_index(opsd_annual.index.year)
opsd_annual.index.name = 'Year'
# Compute the ratio of Wind+Solar to Consumption
opsd_annual['Wind+Solar/Consumption'] = opsd_annual['Wind+Solar'] / opsd_annual['Consumption']
opsd_annual.tail(3)
```

Por último, representemos en un gráfico de barras la cuota de la energía eólica y solar en el consumo anual de electricidad.

```python
# Plot from 2012 onwards, because there is no solar production data in earlier years
ax = opsd_annual.loc[2012:, 'Wind+Solar/Consumption'].plot.bar(color='C0')
ax.set_ylabel('Fraction')
ax.set_ylim(0, 0.3)
ax.set_title('Wind + Solar Share of Annual Electricity Consumption')
plt.xticks(rotation=0);
```

Podemos ver que la producción eólica + solar como porcentaje del consumo anual de electricidad ha ido aumentando desde aproximadamente el 15% en 2012 hasta alrededor del 27% en 2017.

## Ventanas móviles
Las operaciones de ventanas móviles son otra transformación importante para los datos de series temporales. De forma similar al muestreo descendente, las ventanas móviles dividen los datos en ventanas temporales y los datos de cada ventana se agregan con una función como mean(), median(), sum(), etc. Sin embargo, a diferencia del muestreo descendente, en el que los intervalos de tiempo no se solapan y la salida tiene una frecuencia inferior a la de entrada, las ventanas móviles se solapan y "ruedan" a lo largo de la misma frecuencia que los datos, por lo que la serie temporal transformada tiene la misma frecuencia que la serie temporal original.

Por defecto, todos los puntos de datos dentro de una ventana tienen el mismo peso en la agregación, pero esto puede cambiarse especificando tipos de ventana como la gaussiana, la triangular y otras. Aquí nos ceñiremos a la ventana estándar de igual ponderación.

Utilicemos el método rolling() para calcular la media móvil de 7 días de nuestros datos diarios. Usamos el argumento center=True para etiquetar cada ventana en su punto medio, por lo que las ventanas móviles son:

- 2006-01-01 a 2006-01-07 - etiquetada como 2006-01-04
- 2006-01-02 a 2006-01-08 - etiquetadas como 2006-01-05
- 2006-01-03 a 2006-01-09 - etiquetadas como 2006-01-06
- y así sucesivamente...

```python
# Compute the centered 7-day rolling mean
opsd_7d = opsd_daily[data_columns].rolling(7, center=True).mean()
opsd_7d.head(10)
```

Podemos ver que el primer valor de la media móvil que no falta se encuentra en 2006-01-04, porque este es el punto medio de la primera ventana móvil.

Para visualizar las diferencias entre la media móvil y el remuestreo, actualicemos nuestro gráfico anterior de la producción de energía solar de enero a junio de 2017 para incluir la media móvil de 7 días junto con la media semanal de la serie temporal remuestreada y los datos diarios originales.

```python
# Start and end of the date range to extract
start, end = '2017-01', '2017-06'
# Plot daily, weekly resampled, and 7-day rolling mean time series together
fig, ax = plt.subplots()
ax.plot(opsd_daily.loc[start:end, 'Solar'],
marker='.', linestyle='-', linewidth=0.5, label='Daily')
ax.plot(opsd_weekly_mean.loc[start:end, 'Solar'],
marker='o', markersize=8, linestyle='-', label='Weekly Mean Resample')
ax.plot(opsd_7d.loc[start:end, 'Solar'],
marker='.', linestyle='-', label='7-d Rolling Mean')
ax.set_ylabel('Solar Production (GWh)')
ax.legend();
```

Podemos ver que los puntos de datos de la serie temporal media móvil tienen el mismo espaciado que los datos diarios, pero la curva es más suave porque se ha promediado la variabilidad de mayor frecuencia. En la serie temporal media móvil, los máximos y mínimos tienden a alinearse estrechamente con los máximos y mínimos de la serie temporal diaria. Por el contrario, los máximos y mínimos de la serie temporal semanal remuestreada no coinciden tanto con los de la serie temporal diaria, ya que la serie temporal remuestreada tiene una granularidad más gruesa.

## Tendencias
Los datos de las series temporales suelen presentar una variabilidad lenta y gradual, además de una variabilidad de mayor frecuencia, como la estacionalidad y el ruido. Una forma sencilla de visualizar estas tendencias es con medias móviles a diferentes escalas temporales.

Una media móvil tiende a suavizar una serie temporal promediando las variaciones en frecuencias mucho más altas que el tamaño de la ventana y promediando cualquier estacionalidad en una escala temporal igual al tamaño de la ventana. Esto permite explorar las variaciones de menor frecuencia en los datos. Dado que nuestra serie temporal de consumo eléctrico tiene estacionalidad semanal y anual, vamos a analizar las medias móviles en esas dos escalas temporales.

Ya hemos calculado las medias móviles de 7 días, así que ahora vamos a calcular la media móvil de 365 días de nuestros datos OPSD.

```python
# The min_periods=360 argument accounts for a few isolated missing days in the
# wind and solar production time series
opsd_365d = opsd_daily[data_columns].rolling(window=365, center=True, min_periods=360).mean()
```

Vamos a trazar la media móvil de consumo eléctrico de 7 y 365 días, junto con la serie temporal diaria.

```python
# Plot daily, 7-day rolling mean, and 365-day rolling mean time series
fig, ax = plt.subplots()
ax.plot(opsd_daily['Consumption'], marker='.', markersize=2, color='0.6',
linestyle='None', label='Daily')
ax.plot(opsd_7d['Consumption'], linewidth=2, label='7-d Rolling Mean')
ax.plot(opsd_365d['Consumption'], color='0.2', linewidth=3,
label='Trend (365-d Rolling Mean)')
# Set x-ticks to yearly interval and add legend and labels
ax.xaxis.set_major_locator(mdates.YearLocator())
ax.legend()
ax.set_xlabel('Year')
ax.set_ylabel('Consumption (GWh)')
ax.set_title('Trends in Electricity Consumption');
```

Podemos ver que la media móvil de 7 días ha suavizado toda la estacionalidad semanal, preservando al mismo tiempo la estacionalidad anual. La media móvil de 7 días revela que, aunque el consumo de electricidad suele ser mayor en invierno y menor en verano, se produce un descenso drástico durante unas semanas cada invierno, a finales de diciembre y principios de enero, durante las vacaciones.

Si observamos la serie temporal de la media móvil de 365 días, podemos ver que la tendencia a largo plazo del consumo eléctrico es bastante plana, con un par de periodos de consumo anómalamente bajo en torno a 2009 y 2012-2013.

Veamos ahora las tendencias de la producción eólica y solar.

```python
# Plot 365-day rolling mean time series of wind and solar power
fig, ax = plt.subplots()
for nm in ['Wind', 'Solar', 'Wind+Solar']:
    ax.plot(opsd_365d[nm], label=nm)
    # Set x-ticks to yearly interval, adjust y-axis limits, add legend and labels
    ax.xaxis.set_major_locator(mdates.YearLocator())
    ax.set_ylim(0, 400)
    ax.legend()
    ax.set_ylabel('Production (GWh)')
    ax.set_title('Trends in Electricity Production (365-d Rolling Means)');
```

Se observa una pequeña tendencia al alza en la producción de energía solar y una gran tendencia al alza en la producción de energía eólica, ya que Alemania sigue ampliando su capacidad en estos sectores.
