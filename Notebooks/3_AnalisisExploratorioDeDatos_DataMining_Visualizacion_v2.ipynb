# Ingeniería de datos
Arreglar datos, extrayendo características, organizando....

```python
## Removiendo columnas que no sirven
data=data.drop(columns=["S.No."],axis=1)
data.info()
```

```python
data["Modelo"]=data["Name"].str.split(" ").str.get(1)#+data["Name"].str.split(" ").str.get(2)
```

```python
data=data.drop(columns=["Name"])
```

```python
data
```

```python
data["Autonomia"]=data["Mileage"].str.split(" ").str.get(0)
```

```python
data=data.drop(columns=["Mileage"])
```

```python
data["Motor"]=data["Engine"].str.split(" ").str.get(0)
data["Potencia"]=data["Power"].str.split(" ").str.get(0)
```

```python
#Usando diccionarios para cambiar valores de ciertas columnas y hacer nuevas columnas con nuevas características
## Se define un diccionario para cambiar cosas. La sintaxis es dict={"Algo a cambiar 1":"Cambio 1","Algo a cambiar 2":"Cambio 2"}
diccionario={0:"Viejo",1:"Nuevo"}
```

```python
#Luego se mapea (se usa) el diccionario en la column que se desea cambiar valores....
data["Es viejo 2"]=data["Es viejo"].map(diccionario)
```

```python
data.head(20)
```

```python
# Otro ejemplo.....
diccionario2={"Volkswagen":"Gamma Alta","Toyota":"Gamma Alta"}
```
